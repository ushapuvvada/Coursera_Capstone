```python
import wikipedia as wp # library for wiki data
import pandas as pd 
```

```python
#bukding data from url and obtaining data in to pandas library
html = wp.page("List of postal codes of Canada: M").html().encode("UTF-8")
#identifying index of table
df = pd.read_html(html, header = 0)[0]
#The dataframe will consist of three columns: PostalCode, Borough, and Neighborhood
df
```

```python
#Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.
df=df[df['Borough']!='Not assigned']
df
```

```python
#More than one neighborhood can exist in one postal code area.
result = df.groupby(['Postcode','Borough'], sort=False).agg( ', '.join).reset_index()
result
```

#If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough. So for the 9th cell in the table on the Wikipedia page, the value of the Borough and the Neighborhood columns will be Queen's Park.

```python
result.loc[result['Neighbourhood'] =='Not assigned' , 'Neighbourhood'] = result['Borough']
result.head()
```

```python

#In the last cell of your notebook, use the .shape method to print the number of rows of your dataframe.
result.shape
```

```python
#To obtain  the geographical coordinates of each postal code from csv file

!wget -q -O 'Toronto_long_lat_data.csv'  http://cocl.us/Geospatial_data
df_lon_lat = pd.read_csv('Toronto_long_lat_data.csv')
df_lon_lat.head()
```

```python
df_lon_lat.columns=['Postcode','Latitude','Longitude']# csv data 

#Merging data on postal code to map latitude and longitude
Toronto_df = pd.merge(result,
                 df_lon_lat[['Postcode','Latitude', 'Longitude']],
                 on='Postcode')
Toronto_df
```

```python
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans


import folium # map rendering library

print('Libraries imported.')
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="Toronto")
location = geolocator.geocode(address)
latitude_toronto = location.latitude
longitude_toronto = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude_toronto, longitude_toronto))
```

```python
map_toronto = folium.Map(location=[latitude_toronto, longitude_toronto], zoom_start=10)

# add markers to map
for lat, lng, borough, Neighbourhood in zip(Toronto_df['Latitude'], Toronto_df['Longitude'], Toronto_df['Borough'], Toronto_df['Neighbourhood']):
    label = '{}, {}'.format(Neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
